```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib inline
```

```python
df = pd.read_csv('../data/beeline/train.csv.gz') # загружаем данные
len(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.sample(2)
```

```python
# числовые поля
df_num =  df.select_dtypes(include=['float','int'])
len(df_num.columns)
```

```python
cols=4
rows=len(df_num.columns)//cols + ((len(df_num.columns)%cols)>0)
xs=20
ys=xs*(rows//cols+1)
```

```python
# рисуем плотности признаков
df_num.sample(500).plot.kde(
    subplots=True,
    layout=(rows,cols),
    figsize=(xs,ys),
    sharey=False,
    sharex=False,
    grid=True)

plt.show()
```
